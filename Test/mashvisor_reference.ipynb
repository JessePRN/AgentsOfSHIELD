## using http.client to make mashvisor API calls
### requests library works fine, preserving in case there's a need to swap

```python
import http.client 

mashKey = '399480ac-689f-4537-a30a-31d95fa5cd12'

conn = http.client.HTTPSConnection("api.mashvisor.com") 
# https://api.mashvisor.com/v1.1/client/city/neighborhoods/CA/Los%20Angeles

headers = { 
    'accept': "application/json", 
    'x-api-key': mashKey 
} 

conn.request("GET", "/v1.1/client/city/neighborhoods/CA/Los%20Angeles", headers=headers) 

res = conn.getresponse() 
data = res.read() 

print(data.decode("utf-8"))

```
